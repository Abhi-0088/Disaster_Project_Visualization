```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for Damage Estimates in Shillong (in arbitrary units for demonstration)
years = ["2018", "2019", "2020", "2021", "2022"]
infrastructure_damage = [35, 40, 45, 50, 48]  # Damage in infrastructure
building_damage = [25, 30, 28, 35, 33]        # Damage in buildings and homes
environmental_damage = [15, 12, 18, 15, 17]   # Damage to environment and ecology
agricultural_damage = [10, 8, 12, 9, 11]      # Damage to agricultural land

# Stack the damage types for stacked bar chart
ind = np.arange(len(years))
width = 0.6

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(ind, infrastructure_damage, width, label='Infrastructure')
ax.bar(ind, building_damage, width, bottom=infrastructure_damage, label='Buildings & Homes')
ax.bar(ind, environmental_damage, width, bottom=np.array(infrastructure_damage) + np.array(building_damage), label='Environment')
ax.bar(ind, agricultural_damage, width, bottom=np.array(infrastructure_damage) + np.array(building_damage) + np.array(environmental_damage), label='Agricultural Land')

# Labels and Title
ax.set_xlabel("Year")
ax.set_ylabel("Damage Estimates (Units)")
ax.set_title("Damage Estimates due to Landslides in Shillong")
ax.set_xticks(ind)
ax.set_xticklabels(years)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (in millions of currency units)
years = ["2018", "2019", "2020", "2021", "2022"]
repair_costs = [10, 15, 20, 25, 30]           # Cost for repair and rehabilitation
healthcare_costs = [5, 6, 8, 10, 9]           # Cost for healthcare and emergency response
livelihood_losses = [12, 14, 18, 20, 22]      # Loss of livelihoods due to economic disruptions

# Plot multi-line chart
plt.figure(figsize=(10, 6))
plt.plot(years, repair_costs, marker='o', label="Repair Costs")
plt.plot(years, healthcare_costs, marker='o', label="Healthcare Costs")
plt.plot(years, livelihood_losses, marker='o', label="Loss of Livelihoods")

# Labels and Title
plt.xlabel("Year")
plt.ylabel("Economic Loss (in million units)")
plt.title("Economic Loss Progression by Sector in Shillong (2018-2022)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Setting up the data for the clustered bar chart
ind = np.arange(len(years))  # X locations for the groups
width = 0.25  # Width of the bars

# Plot clustered bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(ind - width, repair_costs, width, label="Repair Costs")
bar2 = ax.bar(ind, healthcare_costs, width, label="Healthcare Costs")
bar3 = ax.bar(ind + width, livelihood_losses, width, label="Loss of Livelihoods")

# Labels and Title
ax.set_xlabel("Year")
ax.set_ylabel("Economic Loss (in million units)")
ax.set_title("Economic Loss by Sector in Shillong (2018-2022)")
ax.set_xticks(ind)
ax.set_xticklabels(years)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar chart
hazards = ['Earthquake', 'Landslide', 'Flood', 'Windstorm', 'Drought', 'Tsunami']
impact_ratings = [4.0, 3.33, 3.67, 2.67, 3.0, 2.33]

# Normalize the impact ratings to get colors
norm = plt.Normalize(min(impact_ratings), max(impact_ratings))
colors = plt.cm.viridis(norm(impact_ratings))  # Base color mapping

# Identify the index of the maximum impact rating
max_index = np.argmax(impact_ratings)

# Set the highest magnitude bar to red
colors[max_index] = [1, 0, 0, 1]  # RGBA for red

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(hazards, impact_ratings, color=colors)

# Add titles and labels
plt.title('Impact Ratings of Hazards in Shillong', fontsize=16)
plt.xlabel('Hazard', fontsize=14)
plt.ylabel('Impact Rating', fontsize=14)
plt.ylim(0, 5)  # Set y-axis limits

# Show grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the bar chart
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the radar chart
labels = ['Frequency', 'Magnitude', 'Exposure Time']
hazards = ['Earthquake', 'Landslide', 'Flood', 'Windstorm', 'Drought', 'Tsunami']
data = np.array([
    [4, 5, 3],  # Earthquake
    [3, 4, 3],  # Landslide
    [4, 3, 4],  # Flood
    [3, 3, 2],  # Windstorm
    [2, 3, 4],  # Drought
    [1, 5, 1]   # Tsunami
])

# Number of variables
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot needs to be "closed" for the radar chart
data = np.concatenate((data, data[:, [0]]), axis=1)
angles += angles[:1]

# Create the radar chart
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))

# Plot each hazard
for i in range(len(hazards)):
    ax.fill(angles, data[i], alpha=0.25, label=hazards[i])
    ax.plot(angles, data[i], linewidth=2, linestyle='solid')

# Add labels and title
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Radar Chart for Hazard Analysis', fontsize=16)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Show the radar chart
plt.show()
```
